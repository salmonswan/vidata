```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv(r'../../data/raw/2024_05_28 atlas_antibiotics.csv')
df.head()
```

```python
# Rename the columns in the ATLAS_antibiotics dataset to match the Hilmit structure, except "Age Group"
df.rename(columns={
    'Isolate Id': 'Isolate Number',
    'Source': 'Source of Infection'
    # Species, Family, Gender, Country, State, and Year already match the structure
}, inplace=True)
df.head()
```

```python
# Rename the specified columns
df.rename(columns={
    'Amoxycillin clavulanate': 'Amoxicillin-clavulanate',
    'Trimethoprim sulfa': 'Trimethoprim-sulfamethoxazole'
}, inplace=True)
```

```python
# List of columns that should be kept based on the Hilmit structure (excluding interpretation columns)
columns_to_keep = [
    'Isolate Number', 'Species', 'Family', 'Gender', 'Age Group', 'Country', 'State', 'Year', 'Source of Infection',
    'Amoxicillin-clavulanate', 'Ampicillin', 'Piperacillin tazobactam', 'Ceftriaxone', 'Ceftazidime', 'Cefepime',
    'Imipenem', 'Meropenem', 'Ciprofloxacin', 'Levofloxacin', 'Amikacin', 'Gentamicin', 'Vancomycin', 
    'Azithromycin', 'Clarithromycin', 'Tigecycline', 'Linezolid', 'Trimethoprim-sulfamethoxazole'
]

# Filter the dataframe to keep only the specified columns
df_filtered = df[columns_to_keep]
```

```python
# Drop the Family and State columns from df_filtered
df_filtered.drop(columns=['Family', 'State'], inplace=True)
```

```python
# Define the desired column order for Hilmit
columns_order = [
    'Isolate Number', 'Species', 'Gender', 'Age Group', 'Country', 'Year', 'Source of Infection',
    'Amoxicillin-clavulanate', 'Ampicillin', 'Piperacillin tazobactam', 'Ceftriaxone', 'Ceftazidime', 'Cefepime',
    'Imipenem', 'Meropenem', 'Ciprofloxacin', 'Levofloxacin', 'Amikacin', 'Gentamicin', 'Vancomycin', 
    'Azithromycin', 'Clarithromycin', 'Tigecycline', 'Linezolid', 'Trimethoprim-sulfamethoxazole'
]

# Reorder the dataframe according to the specified column order
df_filtered = df_filtered[columns_order]
```

```python
# Insert 'Data Source' column after 'Isolate Number' with value 'ATLAS'
df_filtered.insert(df_filtered.columns.get_loc('Isolate Number') + 1, 'Data Source', 'ATLAS')
```

```python
# Export df_filtered to a CSV file named 'ATLAS_dataset.csv'
df_filtered.to_csv('../../data/processed/ATLAS_dataset.csv', index=False)
```
